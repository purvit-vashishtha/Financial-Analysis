### Plotting Graph

```python
def show_plot(df, fig_title):
    df.plot(x='Date', figsize = (15,7), linewidth = 3, title = fig_title)
    plt.grid()
    plt.show()
```

```python
show_plot(df,'Raw Stock Prices without Normalisation')
```

```python
def normalize(df):
    x = df.copy()
    for i in x.columns[1:]:
        x[i] = x[i]/x[i][0]
    return x
```

Normalizing Data

```python
normalize(df)
```

```python
show_plot(normalize(df), 'Normalized Raw Stock Prices')
```

### Daily Returns of Individual Stocks

```python
dfe = df['sp500']
print('Daily Returns of S&P 500:')
df_daily_return = dfe.copy()
for j in range(1, len(dfe)):
    df_daily_return[j] = ((dfe[j] - dfe[j-1]) / dfe[j-1]) *100

df_daily_return[0] = 0
df_daily_return
```

```python
dfe = df['AMZN']
print('Daily Returns of Amazon:')
df_daily_return = dfe.copy()
for j in range(1, len(dfe)):
    df_daily_return[j] = ((dfe[j] - dfe[j-1]) / dfe[j-1]) *100

df_daily_return[0] = 0
df_daily_return
```

### Multiple Stocks Daily Return

```python
def daily_return(df):
    df_daily_return = df.copy()
    
    for i in df.columns[1:]:
        
        for j in range(1, len(df)):
            df_daily_return[i][j] = ((df[i][j] - df[i][j-1])/ df[i][j-1]) *100
            
        df_daily_return[i][0] = 0
    
    
    
    return df_daily_return
```

```python
stocks_daily_return = daily_return(df)
stocks_daily_return
```

```python
show_plot(stocks_daily_return,'Stocks Daily Returns')
```

### Correlations Between Daily Returns

```python
cm = stocks_daily_return.drop(columns=['Date']).corr()
```

```python
plt.figure(figsize=(12,12))
sns.heatmap(cm, annot = True)
```

## Asset Allocation

```python
np.random.seed(101)
weights = np.array(np.random.random(9))

weights = weights/ np.sum(weights)

print(weights)
```

```python
df_pf = normalize(df)
df_pf
```

```python
df_pf.columns[1:]
```

```python
for counter, stock in enumerate(df_pf.columns[1:]):
    df_pf[stock] = df_pf[stock] * weights[counter]
    df_pf[stock] = df_pf[stock] * 1000000
    
df_pf
```

```python
df_pf["Portfolio Daily Worth in $"] = df_pf[df_pf != 'Date'].sum(axis = 1)
df_pf
```
